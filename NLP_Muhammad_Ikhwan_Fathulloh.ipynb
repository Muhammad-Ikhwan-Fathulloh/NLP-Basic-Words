```python
# Scripping Web
import requests

# Manipulation String
import string
import re

# Memisahkan layer dari inner HTML
from bs4 import BeautifulSoup

# Natural Language Processing Techniques
import nltk

# Stopwords
from nltk.corpus import stopwords

# Install Library Sastrawi
!pip install Sastrawi
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
```

String and Cleansing

```python
# Request data in web
web = requests.get('https://wartakota.tribunnews.com').text
soup = BeautifulSoup(web)

# Remove tag script and style in html
for i in soup(['script','style']):
    i.decompose()

teks = ''.join(soup.stripped_strings)
print(teks)
```

Stemming

```python
teks = teks.lower()
teks = re.sub(r"\d+","",teks)
teks = teks.translate(str.maketrans("","",string.punctuation))
teks = teks.strip()

factory = StemmerFactory()
stemmer = factory.create_stemmer()
output = stemmer.stem(teks)
print(output)
```

Tokenizing

```python
tokens = [t for t in output.split()]
print(tokens)
```

Stopwords

```python
nltk.download()
clean_tokens = tokens[:]
for token in tokens:
  if token in stopwords.words('indonesian'):
    clean_tokens.remove(token)

freq = nltk.FreqDist(clean_tokens)
for key, val in freq.items():
  print(str(key) + ":" + str(val))

freq.plot(30)
```
